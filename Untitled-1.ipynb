```python
import matplotlib.pyplot as plt
import numpy as np

# 定义四个函数
def g1(x):
    return x**5 - 3*x**3 - 2*x**2 + 2
def g2(x):
    return np.cos(np.sin(x))
def g3(x):
    return x**2 - np.sin(x + 0.15)
def g4(x):
    return x**(x - np.cos(x))

# 定义 x 轴的取值范围
x_range1 = np.linspace(-2.5, 2.5, 1000)
x_range2 = np.linspace(-2.5, 2.5, 1000)
x_range3 = np.linspace(-2.5, 2.5, 1000)
x_range4 = np.linspace(0.8, 1.5, 1000)

# 计算对应的 y 值
y_range1 = g1(x_range1)
y_range2 = g2(x_range2)
y_range3 = g3(x_range3)
y_range4 = g4(x_range4)

# 绘制四个函数图像
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x_range1, y_range1)
axs[0, 0].plot(x_range1, x_range1)
axs[0, 0].set_title('g1')
axs[0, 1].plot(x_range2, y_range2)
axs[0, 1].plot(x_range2, x_range2)
axs[0, 1].set_title('g2')
axs[1, 0].plot(x_range3, y_range3)
axs[1, 0].plot(x_range3, x_range3)
axs[1, 0].set_title('g3')
axs[1, 1].plot(x_range4, y_range4)
axs[1, 1].plot(x_range4, x_range4)
axs[1, 1].set_title('g4')

# 调整子图之间的距离
plt.subplots_adjust(hspace=0.3, wspace=0.3)

# 显示图形
plt.show()
```

```python
def distance1(x):
    return (x - 3) ** 2 + (x ** 2 - 1) ** 2
def distance1_derivative(x):
    return 2 * (x - 3) + 4 * x * (x ** 2 - 1)
def distance1_dderivative(x):
    return 12*x**2-2
def solution():
    x0 = 0.5
    while 1:
        x1 = x0 - distance1_derivative(x0) / distance1_dderivative(x0)
        if abs(x1 - x0) < 1e-12:
            break
        x0 = x1
    return [x1, x1 ** 2]
result = solution()
print("x=", result[0], "y=", result[1])
```
